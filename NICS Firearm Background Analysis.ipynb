```python
# Import Librarie

import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
ax = matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data

df = pd.read_excel(r'C:\Users\MY LAPTOP\Downloads\NICS.xlsx', parse_dates=['month'])
```

```python
df
```

```python
df['Year'] = pd.DatetimeIndex(df['month']).year
```

```python
df['Months'] = df['month'].dt.month_name()
```

```python
df['totals'] = df['totals'].astype(int)
```

```python
## df['ValuesInMillion'] = '$' + (df['totals'].astype(int)/1000000).round(2).astype(str) + 'M'
```

# # Total checks per year and state

```python
BackCheck = df.groupby(['month','Year', 'Months', 'state'])['totals'].sum()
BackCheck
```

```python
BackCheck.to_excel('TCheck.xlsx')
```

# # Converting totals column to millions

```python
df['MillionsAsInt'] = df.apply(lambda row: row['totals'] / 1000000, axis=1)
```

```python
df['MillionsAsString'] = df.apply(lambda row: str(round(row['totals']/ 1000000,2)) + 'M', axis=1)
```

# # Unpivot Firearm columns

```python
FirearmType = pd.melt(df, id_vars=['month',
                                  'state',
                                  'permit',
                                  'permit_recheck',
                                  'admin',
                                  'prepawn_handgun',
                                  'prepawn_long_gun',
                                  'prepawn_other',
                                  'redemption_handgun',
                                  'redemption_long_gun',
                                  'redemption_other',
                                  'returned_handgun',
                                  'returned_long_gun',
                                  'returned_other',
                                  'rentals_handgun',
                                  'rentals_long_gun',
                                  'private_sale_handgun',
                                  'private_sale_long_gun',
                                  'private_sale_other',
                                  'return_to_seller_handgun',
                                  'return_to_seller_long_gun',
                                  'return_to_seller_other',
                                  'totals',
                                  'Year',
                                  'Months',
                                  'MillionsAsInt',
                                  'MillionsAsString'],value_vars=['handgun','long_gun','other','multiple'],var_name='FireArm',value_name='FirearmValue')
```

```python
FirearmType
```

```python
FirearmType.to_excel('FireUn.xlsx')
```

# # Unpivot Transaction Column

```python
Transact = pd.melt(FirearmType, id_vars=['month',
                                  'state',
                                  'FireArm',
                                  'admin',
                                  'FirearmValue',
                                 'totals',
                                  'Year',
                                  'Months',
                                  'MillionsAsInt',
                                  'MillionsAsString'],value_vars=['permit',
                                                                  'permit_recheck',
                                                                  'prepawn_handgun',
                                                                  'prepawn_long_gun',
                                                                  'prepawn_other',
                                                                  'redemption_handgun',
                                                                 'redemption_long_gun',
                                                                 'redemption_other',
                                                                 'returned_handgun',
                                                                 'returned_long_gun',
                                                                 'returned_other',
                                                                 'rentals_handgun',
                                                                 'rentals_long_gun',
                                                                 'private_sale_handgun',
                                                                 'private_sale_long_gun',
                                                                 'private_sale_other',
                                                                 'return_to_seller_handgun',
                                                                 'return_to_seller_long_gun',
                                                                 'return_to_seller_other',
                                                                 ],var_name='Transaction',value_name='TransValue')
```
